1. **What is Simple Linear Regression?**
- Simple linear regression is a statistical method that allows us to summarize and study relationships between two continuous (quantitative) variables:
 - One variable, denoted x, is regarded as the predictor, explanatory, or independent variable.
 - The other variable, denoted y, is regarded as the response, outcome, or dependent variable.
- Because the other terms are used less frequently today, we'll use the "predictor" and "response" terms to refer to the variables encountered in this course. The other terms are mentioned only to make you aware of them should you encounter them. Simple linear regression gets its adjective "simple," because it concerns the study of only one predictor variable. In contrast, multiple linear regression, which we study later in this course, gets its adjective "multiple," because it concerns the study of two or more predictor variables.
- Simple linear regression is a parametric test, meaning that it makes certain assumptions about the data. These assumptions are:
 - Homogeneity of variance (homoscedasticity): the size of the error in our prediction doesn’t change significantly across the values of the independent variable.
 - Independence of observations: the observations in the dataset were collected using statistically valid sampling methods, and there are no hidden relationships among observations.
 - Normality: The data follows a normal distribution.
 - The relationship between the independent and dependent variable is linear: the line of best fit through the data points is a straight line (rather than a curve or some sort of grouping factor).
- The formula for a simple linear regression is:
   
   `y=β0+β1x+ϵ`

 - y is the predicted value of the dependent variable (y) for any given value of the independent variable (x).
 - B0 is the intercept, the predicted value of y when the x is 0.
 - B1 is the regression coefficient – how much we expect y to change as x increases.
 - x is the independent variable ( the variable we expect is influencing y).
 - e is the error of the estimate, or how much variation there is in our estimate of the regression coefficient.

2. **What are the key assumptions of Simple Linear Regression?**

 **Here's a summary of the core assumptions of Simple Linear Regression:**

- Linearity: The relationship between the independent variable (x) and the dependent variable (y) is assumed to be linear. This means that a change in 'x' is associated with a constant and proportional change in 'y'. Violations can lead to inaccurate predictions and misleading inferences about the relationship.
- Independence of Errors: The residuals (or errors – the differences between observed and predicted values) are assumed to be independent of each other. This is especially important in time series data, where consecutive observations shouldn't be correlated. Violations can lead to underestimated standard errors and invalid hypothesis tests.
- Homoscedasticity: The variance of the residuals is assumed to be constant across all levels of the independent variable(s). This means the spread of the residuals should be roughly the same regardless of the value of the independent variable. Violations can lead to biased standard errors, unreliable hypothesis tests, and inaccurate confidence intervals.
- Normality of Residuals: The residuals are assumed to follow a normal (bell-shaped) distribution. This assumption is particularly important for statistical inferences, such as calculating confidence intervals and hypothesis tests. Violations can lead to inaccurate hypothesis tests, invalid confidence intervals, and reduced model performance.

3. **What is heteroscedasticity, and why is it important to address in regression models?**
- Heteroscedasticity refers to situations where the variance of the residuals is unequal over a range of measured values. When running a regression analysis, heteroskedasticity results in an unequal scatter of the residuals (also known as the error term).
- When observing a plot of the residuals, a fan or cone shape indicates the presence of heteroskedasticity. In statistics, heteroskedasticity is seen as a problem because regressions involving ordinary least squares (OLS) assume that the residuals are drawn from a population with constant variance.
- If there is an unequal scatter of residuals, the population used in the regression contains unequal variance, and therefore the analysis results may be invalid.
- **Why it's important to address**

 **Ignoring heteroscedasticity can cause several issues in regression models:**
 - Inefficient Estimates: Ordinary Least Squares (OLS) regression assumes homoscedasticity. With heteroscedasticity, OLS estimators are still unbiased but become less efficient, resulting in larger standard errors and less precise coefficient estimates.
 - Biased Standard Errors: Heteroscedasticity can bias standard errors, making hypothesis tests unreliable. This can lead to incorrect conclusions about the statistical significance of independent variables.
 - Invalid Confidence Intervals: Confidence intervals become unreliable, potentially being too wide or too narrow.
 - Poor Predictions: Predictions from a heteroscedastic model will have high variance, reducing their accuracy and reliability.

4. **What is Multiple Linear Regression?**
- Multiple linear regression (MLR) is a method for estimating how several independent factors together influence a single outcome. It fits a straight-line equation to data points to reveal how each variable contributes when the others are held steady.
- Multiple linear regression (MLR), also known simply as multiple regression, is a statistical technique that uses several explanatory variables to predict the outcome of a response variable.
- The goal of MLR is to model the linear relationship between the explanatory (independent) variables and response (dependent) variables. In essence, multiple regression is the extension of ordinary least-squares (OLS) regression because it involves more than one explanatory variable.
- The equation for a multiple linear regression
  
  Y = β0 + β1X1 + β2X2 + ... + βkXk + ε
  
  Where,
 - Y is the dependent variable being predicted.
 - X1, X2, ... Xk are the independent variables.
 - β0 is the y-intercept.
 - β1, β2, ... βk are the regression coefficients, showing the change in Y for a one-unit change in the corresponding X variable while holding others constant.
 - ε represents the error.
- To find the best-fit line for each independent variable, multiple linear regression calculates three things:
 - The regression coefficients that lead to the smallest overall model error.
 - The t statistic of the overall model.
 - The associated p value (how likely it is that the t statistic would have occurred by chance if the null hypothesis of no relationship between the independent and dependent variables was true).
- It then calculates the t statistic and p value for each regression coefficient in the model.

5. **What is polynomial regression, and how does it differ from linear
regression?**
- Polynomial regression is a type of regression analysis that models the relationship between an independent variable and a dependent variable as an nth-degree polynomial. This means it can fit curved relationships to data, unlike linear regression, which is limited to fitting straight lines.
- Polynomial regression is preferred when the data exhibits a non-linear or curvilinear relationship, meaning a straight line won't accurately capture the underlying trend. Examples include modeling the growth rate of bacteria or predicting crop yield based on environmental factors like temperature and rainfall.

 **Key differences**

 - Relationship modeling: Linear regression assumes and fits a straight-line relationship between variables, according to GeeksforGeeks. Polynomial regression, on the other hand, can model non-linear relationships by fitting a curved line (a polynomial equation) to the data.
 - Model complexity: Linear regression models are simpler and easier to interpret, with fewer parameters. Polynomial regression models are more complex due to the inclusion of higher-degree polynomial terms.
 - Overfitting: Linear regression is less prone to overfitting but may underfit complex, non-linear data. Polynomial regression, especially with a high polynomial degree, is more susceptible to overfitting, where it fits the noise in the data rather than the underlying patterns


6. **Implement a Python program to fit a Simple Linear Regression model to
the following sample data:**

 ● X = [1, 2, 3, 4, 5]

 ● Y = [2.1, 4.3, 6.1, 7.9, 10.2]

 **Plot the regression line over the data points.**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Reshape for sklearn
y = np.array([2.1, 4.3, 6.1, 7.9, 10.2])

# Create and fit the model
model = LinearRegression()
model.fit(X, y)

# Predictions
y_pred = model.predict(X)

# Plot the data points and regression line
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, y_pred, color='red', label='Regression Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Simple Linear Regression')
plt.legend()
plt.show()

# Display slope and intercept
print("Slope (Coefficient):", model.coef_[0])
print("Intercept:", model.intercept_)
```

7. **Fit a Multiple Linear Regression model on this sample data**:

 ● **Area** = [1200, 1500, 1800, 2000]

 ● **Rooms** = [2, 3, 3, 4]

 ● **Price** = [250000, 300000, 320000, 370000]

 **Check for multicollinearity using VIF and report the results.**

```python
# Import libraries
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from statsmodels.stats.outliers_influence import variance_inflation_factor
import statsmodels.api as sm

# Sample data
Area = [1200, 1500, 1800, 2000]
Rooms = [2, 3, 3, 4]
Price = [250000, 300000, 320000, 370000]

# Create DataFrame
df = pd.DataFrame({
    'Area': Area,
    'Rooms': Rooms,
    'Price': Price
})

# Features (X) and Target (y)
X = df[['Area', 'Rooms']]
y = df['Price']

# Fit Multiple Linear Regression model
model = LinearRegression()
model.fit(X, y)

# Predictions
y_pred = model.predict(X)

# Results
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
print("Predicted Prices:", y_pred)

# ---------------------------
# Check Multicollinearity using VIF
# ---------------------------

# Add constant for statsmodels
X_with_const = sm.add_constant(X)

# Calculate VIF for each feature
vif_data = pd.DataFrame()
vif_data["Feature"] = X_with_const.columns
vif_data["VIF"] = [variance_inflation_factor(X_with_const.values, i)
                   for i in range(X_with_const.shape[1])]

print("\nVariance Inflation Factor (VIF):")
print(vif_data)
```

8. **Implement polynomial regression on the following data:**

 ● X = [1, 2, 3, 4, 5]

 ● Y = [2.2, 4.8, 7.5, 11.2, 14.7]

 **Fit a 2nd-degree polynomial and plot the resulting curve.**

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Sample Data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2.2, 4.8, 7.5, 11.2, 14.7])

# Create Polynomial Features (degree=2)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Fit Polynomial Regression Model
model = LinearRegression()
model.fit(X_poly, y)

# Predict values
y_pred = model.predict(X_poly)

# Plot the results
plt.scatter(X, y, color='blue', label="Actual Data")
plt.plot(X, y_pred, color='red', label="2nd Degree Polynomial Curve")
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Polynomial Regression (Degree 2)')
plt.legend()
plt.show()

# Display coefficients
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
```

9. **Create a residuals plot for a regression model trained on this data:**

 ● X = [10, 20, 30, 40, 50]

 ● Y = [15, 35, 40, 50, 65]

 **Assess heteroscedasticity by examining the spread of residuals.**

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Data
X = np.array([10, 20, 30, 40, 50]).reshape(-1, 1)
y = np.array([15, 35, 40, 50, 65])

# Train Linear Regression model
model = LinearRegression()
model.fit(X, y)

# Predictions
y_pred = model.predict(X)

# Residuals
residuals = y - y_pred

# Plot Residuals
plt.scatter(X, residuals, color='blue')
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel('X')
plt.ylabel('Residuals')
plt.title('Residuals Plot')
plt.show()

# Print residuals for clarity
print("Residuals:", residuals)
```

10. **Imagine you are a data scientist working for a real estate company. You
need to predict house prices using features like area, number of rooms, and location. However, you detect heteroscedasticity and multicollinearity in your regressionmodel. Explain the steps you would take to address these issues and ensure a robust model.**
- **Heteroscedasticity** : This means that the variance of the errors in the regression model is not constant across all levels of the independent variables. In simpler terms, the spread of the residuals (the differences between predicted and actual house prices) changes as the independent variables (like area or number of rooms) change. A residual plot showing a funnel or cone shape is a classic sign of heteroscedasticity.

 **Detecting the issues**

 - **Visual Inspection:** Plotting the residuals against the fitted values or independent variables is often the first step. A fan or cone shape indicates heteroscedasticity.

 - **Statistical Tests:** Formal tests like the Breusch-Pagan test or the White test can be used to statistically confirm the presence of heteroscedasticity. The Breusch-Pagan test checks if the variance of the residuals is related to the independent variables. The White test is a more robust alternative, testing the relationship with the independent variables, their squares, and interaction terms.

 **Addressing heteroscedasticity**

 - **Data Transformations:** Applying transformations like the logarithm or square root to the dependent variable (house price) can help stabilize the variance. However, this can make the results less intuitive to interpret.

 - **Robust Standard Errors:** This approach adjusts the standard errors of the regression coefficients to account for heteroscedasticity without altering the model itself. Robust standard errors (e.g., Huber-White standard errors) are particularly useful when the form of heteroscedasticity is unknown or complex.

 - **Weighted Least Squares (WLS):** WLS assigns weights to observations inversely proportional to the variance of their error terms, downweighting observations with higher variances. This minimizes the sum of weighted squared residuals, addressing the heteroscedasticity issue and potentially leading to more efficient coefficient estimates.

- **Multicollinearity:** This occurs when two or more independent variables in the model are highly correlated with each other. For example, if "area" and "number of rooms" are very strongly correlated, it becomes difficult for the model to isolate the individual impact of each on house price. High Variance Inflation Factor (VIF) scores (e.g., above 5 or 10) are a common indicator of multicollinearity.

 **Detecting the issues**

 - **Correlation Matrix:** Calculating the correlation coefficients between pairs of independent variables can reveal strong linear relationships.

 - **Variance Inflation Factor (VIF):** VIF quantifies how much the variance of an estimated regression coefficient is inflated due to multicollinearity. VIF values greater than 5 or 10 generally suggest that multicollinearity is a problem.

 - **Condition Index:** This examines the relationships between eigenvalues of the correlation matrix, with values above 10 indicating moderate multicollinearity and above 30 indicating severe multicollinearity.

 **Addressing multicollinearity**

 - **Remove Highly Correlated Predictors:** Identifying and removing one of the highly correlated variables, especially if they are redundant or less important for prediction, can reduce multicollinearity. However, careful consideration is needed to avoid losing valuable information.

 - **Combine Variables:** If variables are highly correlated and seem to represent a similar underlying concept, they can be combined into a single predictor using techniques like Principal Component Analysis (PCA). PCA creates a new set of uncorrelated components that capture most of the data's variation. However, this can make the interpretation of the new components challenging.

 - **Ridge Regression and Lasso Regression:** These are regularization techniques that add penalty terms to the regression model, shrinking the coefficients of correlated variables. Ridge regression shrinks coefficients towards zero but doesn't eliminate them entirely, while Lasso regression can set some coefficients to zero, performing feature selection.

- **Building a robust model**

 - Iterative Approach: Addressing these issues may require an iterative process of testing, applying remedies, and re-evaluating the model's performance and robustness.

 - Consider Model Goals: The best approach depends on the primary goal of the model. If the main aim is prediction, high multicollinearity might be acceptable if the overall predictive power is good. However, for understanding the individual impact of variables, addressing multicollinearity becomes more crucial.

 - Domain Expertise: Leveraging domain knowledge in real estate is critical. Understanding the relationships between features and house prices can guide decisions on variable selection, transformations, and interpretation of the results.

 - Validate the Model: After implementing solutions, it's essential to rigorously validate the model's performance on unseen data to ensure its robustness and generalization ability.
